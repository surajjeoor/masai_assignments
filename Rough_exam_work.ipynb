```python
import pandas as pd
import numpy as np

records = [
    {"country": "India", "continent": "Asia", "year": 2007, "population": 1110000000, "gdpPercap": 2452, "lifeExp": 64.7},
    {"country": "China", "continent": "Asia", "year": 2007, "population": 1318000000, "gdpPercap": 4959, "lifeExp": 73.0},
    {"country": "Japan", "continent": "Asia", "year": 2007, "population": 127000000, "gdpPercap": 31656, "lifeExp": None},
    {"country": "France", "continent": "Europe", "year": 2007, "population": 61000000, "gdpPercap": 30470, "lifeExp": 80.7},
    {"country": "Germany", "continent": "Europe", "year": 2007, "population": 82400000, "gdpPercap": 32170, "lifeExp": 79.4},
    {"country": "France", "continent": "Europe", "year": 2007, "population": 61000000, "gdpPercap": 30470, "lifeExp": 80.7},
    {"country": "Brazil", "continent": "Americas", "year": 2007, "population": 190000000, "gdpPercap": 9065, "lifeExp": 72.4},
    {"country": "Canada", "continent": "Americas", "year": 2007, "population": 33300000, "gdpPercap": 36319, "lifeExp": 80.6}
]
```

```python
pd_recs=pd.DataFrame(records)
```

```python
pd_recs
```

```python
pd_recs=pd_recs.drop_duplicates(keep='first')
```

```python
avg_life_exp=pd_recs['lifeExp'].mean()
```

```python
pd_recs['lifeExp']=pd_recs['lifeExp'].fillna(avg_life_exp)
```

```python
pd_recs['gdp']=pd_recs['population']*pd_recs['gdpPercap']
```

```python
pd_recs
```

```python
#for the year 2007 create a continentwise summary containing:
#GDP per continent
#Average life expectancy

pd_recs_2007=pd_recs[pd_recs['year']==2007]
pd_recs_2007=pd_recs_2007.groupby('continent').agg({'gdp':'sum','lifeExp':'mean'})
pd_recs_2007
```

```python
#sorting continentwise summary in the descending order
pd_recs_2007.sort_values(by=['continent'],ascending=[True],inplace=True)
pd_recs_2007
```

```python
def get_status(status_code):
  if status_code==200:
    return f"{status_code} : success"
  elif status_code==404:
    return f"{status_code} : Not found"
  elif status_code==500:
    return f"{status_code} : Internal server error"
  else:
    return "unknown"
```

```python
#top 10 cryptocurrencies data in INR
response=requests.get("https://api.coingecko.com/api/v3/coins/markets",
                      params={
                          'vs_currency':'inr',
                          'order':'market_cap_desc',
                          'per_page':10,
                          'page':1
                      })
if response.status_code == 200:
  crypto=response.json()
  df_crypto=pd.DataFrame(crypto)[['name','symbol','current_price','market_cap','price_change_percentage_24h','total_volume']]
  df_crypto.columns=['Name','Symbol','Price(INR)','Market Cap','24h change %','Volume 24h']
  df_crypto['Price(INR)']=df_crypto['Price(INR)'].apply(lambda x: f"Rs.{x:,.2f}")
  df_crypto['Market Cap']=df_crypto['Market Cap'].apply(lambda x: f"Rs.{x/1e7:,.0f} Cr")
  df_crypto['24h change %']=df_crypto['24h change %'].apply(lambda x: f"{x:,.2f}%")
  df_crypto['Volume 24h']=df_crypto['Volume 24h'].apply(lambda x: f"Rs.{x/1e7:,.0f} Cr")
  df_crypto.index=range(1,len(df_crypto)+1)
  df_crypto.index.name='Rank'
  df_crypto
else:
  print(f"error:{response.status_code}")
```

```python
#Scrape the books from books.toscrape.com
resp=requests.get("https://books.toscrape.com")
soup=BeautifulSoup(resp.text,"html.parser")
books=soup.find_all("article",class_="product_pod")
all_books=[]

for book in books:
  title=book.find("h3").find("a").get("title")
  price=book.find("p",class_="price_color").text
  price_clean = price.replace("£", "").replace("Â", "").strip()
  all_books.append({"title":title,"price":price_clean})

df_books=pd.DataFrame(all_books)
df_books.index=range(1,len(df_books)+1)
print(f"total {len(df_books)} books scraped.")
df_books.head(10)
```
